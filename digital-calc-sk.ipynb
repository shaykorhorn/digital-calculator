# Digital Calculator

```python
import math

class Calculator():
    def __init__(self, input1, input2):
        self.input1 = input1
        self.input2 = input2
    
    def addition(self, num1, num2):
        print("Addition: ", num1 + num2)
    
    def subtraction(self, num1, num2):
        print("Subtraction: ", num1 - num2)
    
    def multiplication(self, num1, num2):
        print("Multiplication: ", num1 * num2)
        
    def division(self, num1, num2):
        if num2 == 0:
            print("Infinite.")
        else:
            print("Division: ", num1 / num2)
        
    def sqrt(self, num1):
        print("Sqrt: ", math.sqrt(num1))
    
    def cbrt(self, num1):
        print("Cbrt: ", math.cbrt(num1))
    
    def acos(self, num1):
        print("Acos: ", math.acos(num1))
    
    def tanh(self, num1):
        print("Tanh: ", math.tanh(num1))
        
    def sinh(self, num1):
        print("Sinh: ", math.sinh(num1))
    
    def degrees(self, num1):
        print("Degrees: ", math.degrees(num1))
    
    def radians(self, num1):
        print("Radians: ", math.radians(num1))
        
    def remainder(self, num1):
        print("Remainder: ", math.remainder(num1))
        

calculator = Calculator(0, 0)

while True:
    print("Select a number: ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Sqrt")
    print("6. Cbrt")
    print("7. Acos")
    print("8. Tanh")
    print("9. Sinh")
    print("10. Degrees")
    print("11. Radians")
    print("12. Remainder")
    print("13. Exit the digital calculator")
    choice = int(input("Enter your choice (1-13): "))
    
    if choice == 1:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        calculator.addition(num1, num2)
    
    elif choice == 2:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        calculator.subtraction(num1, num2)
    
    elif choice == 3:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        calculator.multiplication(num1, num2)
    
    elif choice == 4:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        calculator.division(num1, num2)
        
    elif choice == 5:
        num1 = int(input("Enter the number to find the sqrt: "))
        calculator.sqrt(num1)
        
    elif choice == 6:
        num1 = int(input("Enter the number to find the cbrt: "))
        calculator.cbrt(num1)
    
    elif choice == 7:
        num1 = int(input("Enter the number to find the acos: "))
        calculator.acos(num1)
    
    elif choice == 8:
        num1 = int(input("Enter the number to find the tanh: "))
        calculator.tanh(num1)
        
    elif choice == 9:
        num1 = int(input("Enter the number to find the sinh: "))
        calculator.sinh(num1)
        
    elif choice == 10:
        num1 = int(input("Enter the number to find the degrees: "))
        calculator.degrees(num1)
    
    elif choice == 11:
        num1 = int(input("Enter the number to find the radians: "))
        calculator.radians(num1)
    
    elif choice == 12:
        num1 = int(input("Enter the number to find the remainder: "))
        calculator.remainder(num1)
        
    elif choice == 13:
        break
    
    else:
        print("Not the correct number, please enter a number in between 1 to 13.")
```
